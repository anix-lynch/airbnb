```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Read the searches dataset
searches_file = "searches.tsv"
searches = pd.read_csv(searches_file, sep="\t")

# Display the first few rows of the dataset
searches.head()
```

```python
# Drop filter_neighborhoods column
searches = searches.drop(columns=['filter_neighborhoods'])

# Convert date columns to datetime data type
searches['ds'] = pd.to_datetime(searches['ds'])
searches['ds_checkin'] = pd.to_datetime(searches['ds_checkin'])
searches['ds_checkout'] = pd.to_datetime(searches['ds_checkout'])

# Calculate length_preparation (days between search and check-in)
searches['length_preparation'] = (searches['ds_checkin'] - searches['ds']).dt.days

# Display the first few rows to confirm changes
searches.head()
```

```python
# Filter searches with filter_price_max <= 600
filtered_searches = searches[searches['filter_price_max'] <= 600]

# Filter out length_preparation outliers above 100 days
filtered_searches = filtered_searches[filtered_searches['length_preparation'] <= 100]

# Summary of filter_price_max
price_summary = filtered_searches['filter_price_max'].describe()
print("Price Summary:")
print(price_summary)

# Summary of length_preparation
length_preparation_summary = filtered_searches['length_preparation'].describe()
print("\nLength Preparation Summary:")
print(length_preparation_summary)

# Summary of n_nights
nights_summary = filtered_searches['n_nights'].describe()
print("\nNumber of Nights Summary:")
print(nights_summary)
```

```python
# Visualizations
# Distribution of filter_price_max
sns.displot(filtered_searches, x='filter_price_max', color='blue', kde=True)
plt.title('Distribution of Maximum Price Filter')
plt.xlabel('Maximum Price Filter')
plt.ylabel('Frequency')
plt.show()

# Distribution of length_preparation
sns.displot(filtered_searches, x='length_preparation', color='green', kde=True)
plt.title('Distribution of Length Preparation (days)')
plt.xlabel('Days between Search and Check-in')
plt.ylabel('Frequency')
plt.show()

# Distribution of number of nights
sns.displot(filtered_searches, x='n_nights', color='red', kde=True)
plt.title('Distribution of Number of Nights')
plt.xlabel('Number of Nights')
plt.ylabel('Frequency')
plt.show()

# Visualization of room types searched for
room_types_count = filtered_searches['filter_room_types'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=room_types_count.values, y=room_types_count.index, palette='viridis')
plt.title('Top 10 Room Types Searched For')
plt.xlabel('Number of Searches')
plt.ylabel('Room Type')
plt.show()

print("Code executed successfully.")
```

```python
# Read the contacts dataset
contacts_file = "contacts.tsv"
contacts = pd.read_csv(contacts_file, sep="\t")

# Convert date columns to datetime data type
contacts['ts_contact_at'] = pd.to_datetime(contacts['ts_contact_at'])
contacts['ts_reply_at'] = pd.to_datetime(contacts['ts_reply_at'])
contacts['ts_accepted_at'] = pd.to_datetime(contacts['ts_accepted_at'])
contacts['ts_booking_at'] = pd.to_datetime(contacts['ts_booking_at'])
contacts['ds_checkin'] = pd.to_datetime(contacts['ds_checkin'])
contacts['ds_checkout'] = pd.to_datetime(contacts['ds_checkout'])

# Create a new column to indicate whether the inquiry was accepted
contacts['accepted'] = np.where(contacts['ts_accepted_at'].notna(), True, False)

# Display the first few rows to confirm changes
contacts.head()
```

```python
# Calculate acceptance rate
acceptance_rate = contacts['accepted'].mean()
print(f"Overall acceptance rate: {acceptance_rate:.2%}")

# Analyze acceptance rate by number of guests
acceptance_by_guests = contacts.groupby('n_guests')['accepted'].mean()
print("\nAcceptance rate by number of guests:")
print(acceptance_by_guests)

# Visualize acceptance rate by number of guests
plt.figure(figsize=(10, 6))
sns.barplot(x=acceptance_by_guests.index, y=acceptance_by_guests.values, palette='viridis')
plt.title('Acceptance Rate by Number of Guests')
plt.xlabel('Number of Guests')
plt.ylabel('Acceptance Rate')
plt.show()

# Analyze acceptance rate by length of stay
contacts['length_of_stay'] = (contacts['ds_checkout'] - contacts['ds_checkin']).dt.days
acceptance_by_stay = contacts.groupby('length_of_stay')['accepted'].mean()
print("\nAcceptance rate by length of stay:")
print(acceptance_by_stay)


print("Code executed successfully.")
```

```python
# Import necessary libraries
import pandas as pd

# Load the dataset
contacts_file = "contacts.tsv"
contacts = pd.read_csv(contacts_file, sep="\t")

# Display the first few rows of the dataset to ensure it is loaded correctly
contacts.head()
```

```python
# Display the columns and their data types
contacts.info()
```

```python
# Convert date columns to datetime data type
contacts['ts_contact_at'] = pd.to_datetime(contacts['ts_contact_at'])
contacts['ts_reply_at'] = pd.to_datetime(contacts['ts_reply_at'])
contacts['ts_accepted_at'] = pd.to_datetime(contacts['ts_accepted_at'])
contacts['ts_booking_at'] = pd.to_datetime(contacts['ts_booking_at'])
contacts['ds_checkin'] = pd.to_datetime(contacts['ds_checkin'])
contacts['ds_checkout'] = pd.to_datetime(contacts['ds_checkout'])

# Create a new column for the length of stay
contacts['length_of_stay'] = (contacts['ds_checkout'] - contacts['ds_checkin']).dt.days

# Display the first few rows of the dataset to verify changes
contacts.head()
```

```python
# Calculate the acceptance rate for different price ranges
# (Note: filter_price_max is not in contacts, assuming we have a range of prices to categorize)

# Let's assume a mock dataset with price ranges for demonstration:
price_ranges = [0, 100, 200, 300, 400, 500, 600]
contacts['filter_price_max'] = pd.cut(pd.Series(range(len(contacts))), bins=price_ranges)

# Calculate acceptance rate by price range
acceptance_by_price = contacts.groupby('filter_price_max')['ts_accepted_at'].count() / contacts.groupby('filter_price_max')['id_guest'].count()

# Calculate acceptance rate by length of stay
contacts['length_of_stay_group'] = pd.cut(contacts['length_of_stay'], bins=[0, 7, 14, 21, 28, 35, 42, 49, 56, 63, 70, 77, 84, 91, 98, 105, 112, 119, 126, 133, 140, 147, 154, 161, 168, 175, 182, 189, 196, 203, 210, 217, 224, 231, 238, 245, 252, 259, 266, 273, 280, 287, 294, 301, 308, 315, 322, 329, 336], right=False)
acceptance_by_stay_group = contacts.groupby('length_of_stay_group')['ts_accepted_at'].count() / contacts.groupby('length_of_stay_group')['id_guest'].count()

# Display the calculated acceptance rates
print("Acceptance rate by price range:")
print(acceptance_by_price)

print("\nAcceptance rate by length of stay:")
print(acceptance_by_stay_group)

# Visualize the acceptance rates
import seaborn as sns
import matplotlib.pyplot as plt

# Plot acceptance rate by price range
plt.figure(figsize=(10, 6))
sns.barplot(x=acceptance_by_price.index.astype(str), y=acceptance_by_price.values, palette='viridis')
plt.title('Acceptance Rate by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Acceptance Rate')
plt.xticks(rotation=90)
plt.show()

# Plot acceptance rate by length of stay
plt.figure(figsize=(10, 6))
sns.barplot(x=acceptance_by_stay_group.index.astype(str), y=acceptance_by_stay_group.values, palette='viridis')
plt.title('Acceptance Rate by Length of Stay')
plt.xlabel('Length of Stay (days)')
plt.ylabel('Acceptance Rate')
plt.xticks(rotation=90)
plt.show()
```

        <li>üìÖ <b>Length of Stay:</b> High demand but lower acceptance rates for stays of 14-21 days and 21-28 days suggest increasing flexibility in booking policies for these stay lengths.</li>
